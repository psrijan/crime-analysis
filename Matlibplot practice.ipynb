## Working with Multiple figures

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0 , 10 , 1 )
t2 = np.arange(0 , 20 , 2 )

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1) , 'bo' , t2 , f(t2) , 'k' )
plt.xlabel('People', fontsize=14, color='red')
plt.ylabel('Time' , fontsize=12, color = 'blue')


plt.subplot(212)

plt.plot(t2, np.cos(2*np.pi*t2) , 'r--')
plt.xlabel('People', fontsize=14, color='red')
plt.ylabel('Time' , fontsize=12, color = 'blue')
plt.show()
```

## Annotating Text


```python
ax = plt.subplot(111)
t = np.arange(0, 5 , .01)
s = np.cos(2*np.pi*t)
line = plt.plot(t , s , lw= 2)
plt.annotate('local max' , xy = (2,1) , xytext = (3,1.5) , arrowprops = dict(facecolor='black' , shrink = .05),)

plt.ylim(-2, 2)
plt.grid(True)
plt.show()
```
